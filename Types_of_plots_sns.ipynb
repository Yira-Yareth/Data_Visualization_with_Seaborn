## Making a comparison of plotting with pandas and seaborn

```python
#Importing packages
import io 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Reading csv
df=pd.read_csv("datasets//schoolimprovement2010grants.csv")
print(df.info())
```

```python
# Display pandas histogram
df['Award_Amount'].plot.hist()
plt.show()

# Clear out the pandas histogram
plt.clf()
```

```python
#Use Seaborn's distplot() function to plot a distribution plot of the same column

# Display a Seaborn distplot
sns.distplot(df['Award_Amount'])
plt.show()
# Clear the distplot
plt.clf()
```

#### Notice how the pandas and Seaborn plots are complementary. They both show the distribution of data in different formats

The distplot() function will return a Kernel Density Estimate (KDE) by default. The KDE helps to smooth the distribution and is a useful way to look at the data

```python
# Create a distplot with KDE funcion as true
sns.distplot(df['Award_Amount'],
             kde=True,
             bins=20)

# Display the plot
plt.show()
```

```python
# Create a distplot with KDE funcion as false
sns.distplot(df['Award_Amount'],
             kde=False,
             bins=20)

# Display the plot
plt.show()
```

The distplot() function can be configured with multiple different arguments. By disabling the KDE output, you have created a histogram.

## The process of creating a visualization and updating it in an incremental fashion is a useful and common approach to look at data from multiple perspectives.

```python
# Create a distplot of the Award Amount
sns.distplot(df['Award_Amount'],
             hist=False,
             #Add a rug plot above the x axis.
             rug=True,
             kde_kws={'shade':True})

# Plot the results
plt.show()
```

Looking at this distplot, we can infer based on the visualization; there are a large group of award amounts < $400K.

Since we are using lmplot() now, we can look at the more complex interactions of data. This data set includes geographic information by state and area. It might be interesting to see if there is a difference in relationships based on the Region of the country

lmplot() allows us to facet the data across multiple rows and columns. In the previous plot, the multiple lines were difficult to read in one plot. We can try creating multiple plots by Region to see if that is a more useful visualization

```python
# Create a regression plot with multiple rows
sns.lmplot(data=fivete,
           x="insurance_losses",
           y="premiums",
           row="Region")

# Show the plot
plt.show()
```
